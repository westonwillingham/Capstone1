```python
columns = ['Alcohol','Amphet', 'Amyl', 'Benzos', 'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack',
           'Ecstasy', 'Heroin', 'Ketamine', 'Legalh', 'LSD', 'Meth', 'Mushrooms','Nicotine', 'Semer', 'VSA']
fig, axes = plt.subplots(5,3, figsize = (25,25))
fig.suptitle("Count of Different Classes Vs Drug",fontsize=14)
k=0
for i in range(5):
    for j in range(3):
        sns.countplot(x=columns[k], data=df,ax=axes[i][j])
        k+=1

plt.tight_layout()
plt.show()
```

```python
columns = ['Mushrooms','Nicotine', 'Semer', 'VSA']
fig, axes = plt.subplots(2,2, figsize = (25,25))
fig.suptitle("Count of Different Classes Vs Drug",fontsize=14)
k=0
for i in range(2):
    for j in range(2):
        sns.countplot(x=columns[k], data=df,ax=axes[i][j])
        k+=1

plt.tight_layout()
plt.show()
```

```python
columns = ['Mushrooms','Nicotine', 'Semer', 'VSA']
fig, axes = plt.subplots(figsize = (15,15))
fig.suptitle("Count of Different Classes Vs Drug",fontsize=14)

sns.countplot(x=df["Mushrooms"], data = df)

plt.tight_layout()
plt.show()
```
